# Scalling

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
df.columns
```

```python

numerical_trf = Pipeline(
    steps=[
        ("numScaler",StandardScaler()),
    ]
)
categorical_trf = Pipeline(
    steps=[
        ('ohe_encod',OneHotEncoder(handle_unknown='ignore'))
    ]
)
from sklearn.compose import ColumnTransformer

num_cols = ['age', 'weight(kg)', 'height(m)','activity_level']

preprocessor = ColumnTransformer(
    transformers = [
        ('num_stand',numerical_trf,num_cols),
        ('cat_ohe',categorical_trf,['gender'])
    ],remainder='passthrough'
)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X = df[['age', 'weight(kg)','gender', 'height(m)','activity_level']]
y = df['BMI']
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_train_trf = preprocessor.fit_transform(X_train)
X_test_trf = preprocessor.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train_trf,y_train)
```

```python
y_pred = lr.predict(X_test_trf)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import joblib
joblib.dump(lr,'BMI_Predictor')
```
